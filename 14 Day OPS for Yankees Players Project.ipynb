```python
#Select all rows that match Team name with 'New York', for some reason both Mets and Yankees are listed as NY
newyork_df = data.loc[data['Tm'] == 'New York']
newyork_df.head(20)
```

```python
#select all players and their OPS that played for New York team and asorted it descending 
newyork_df[['Name', 'OPS']].sort_values('OPS', ascending = False)
```

```python
#select the rows with Yankees players I'm interested in looking at by locating their ID numbers (probably a better way of doing this)
yankees_df=newyork_df.loc[[217, 173, 237, 437, 79, 124, 464, 462, 256, 158, 394, 244],['Name', 'OPS']].sort_values('OPS', ascending = False)
yankees_df
```
